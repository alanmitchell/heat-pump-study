```python
# Left out points > 40 F due conversation with Bruce; operating 
# conditions were not the same between coheat and heat pump test.
# And, operation regime of the heat pump was variable.
dt = """7.465618860510805, 1.887640449438205
12.416502946954814, 3.370786516853933
15.02946954813359, 2.966292134831459
17.36738703339882, 2.764044943820224
19.842829076620824, 2.764044943820224
22.318271119842827, 2.966292134831459
24.931237721021617, 2.8314606741573023
27.406679764243613, 3.1685393258426977
29.882121807465616, 2.8988764044943807
32.495088408644406, 3.3033707865168545
34.9705304518664, 3.9101123595505634
37.445972495088405, 4.516853932584272
39.921414538310415, 5.123595505617978
"""
temperature = []
cop = []
for lin in dt.splitlines():
    t, c = lin.split(',')
    temperature.append(float(t))
    cop.append(float(c))
df = pd.DataFrame({'temperature': temperature, 'cop': cop})
df.plot(x='temperature', y='cop', kind='scatter')
```

```python
df['source'] = 'Harley Home'
df['field_study'] = True
df['hspf'] = 12.0   
df.to_pickle('temp_cop_harley.pkl')
```
